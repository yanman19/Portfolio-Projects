```python
def make_predictions_with_assumptions(model, assumptions, date_column, future_periods, frequency):
    """
    model: Trained Prophet model
    assumptions: Dictionary containing assumptions for each variable
    date_column: Name of the date column
    future_periods: Number of periods to make predictions for
    frequency: Frequency of the predictions ('D' for daily, 'M' for monthly, etc.)
    """
    # Create a future DataFrame
    future = model.make_future_dataframe(periods=future_periods, freq=frequency)

    # Add the assumptions to the future DataFrame
    for var in assumptions:
        if var in future.columns:
            raise ValueError(f"Variable {var} conflicts with existing columns in the DataFrame.")
        future[var] = assumptions[var]

    # Make predictions
    forecast = model.predict(future)

    # Return the predictions
    return forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]

# Example usage
# assumptions = {'additional_regressor1': [value1, value2, ...], 'additional_regressor2': [value1, value2, ...]}
# predictions = make_predictions_with_assumptions(model, assumptions, 'ds', future_periods=365, frequency='D')
```

```python
from prophet import Prophet
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error

def train_prophet_model(df, input_month, date_column, target_column):
    # Convert the date column to datetime
    df[date_column] = pd.to_datetime(df[date_column])

    # Filter training data to include only the same month across years, excluding the most recent year
    training_data = df[df[date_column].dt.month == input_month]
    training_data = training_data[training_data[date_column].dt.year < training_data[date_column].dt.year.max()]

    # Test data is the most recent year
    test_data = df[df[date_column].dt.year == df[date_column].dt.year.max()]

    # Identify and drop columns with more than 75% missing values
    cols_to_drop = df.columns[df.isnull().mean() > 0.75]
    df = df.drop(columns=cols_to_drop)

    # Identify regressor columns (all columns except date and target)
    regressor_columns = [col for col in df.columns if col not in [date_column, target_column]]

    # Handle remaining missing values in regressors
    for col in regressor_columns:
        df[col].fillna(df[col].mean(), inplace=True)

    # Prepare data for Prophet
    training_data = training_data.rename(columns={date_column: 'ds', target_column: 'y'})
    test_data = test_data.rename(columns={date_column: 'ds', target_column: 'y'})

    # Initialize the model
    model = Prophet()

    # Add regressor columns to the model
    for col in regressor_columns:
        model.add_regressor(col)

    # Fit the model
    model.fit(training_data)

    # Prepare test data with regressors
    test_data_for_prediction = test_data[['ds'] + regressor_columns]

    # Make predictions
    forecast = model.predict(test_data_for_prediction)

    # Calculate metrics
    mse = mean_squared_error(test_data['y'], forecast['yhat'])
    mae = mean_absolute_error(test_data['y'], forecast['yhat'])

    # Print metrics
    print(f"Mean Squared Error: {mse}")
    print(f"Mean Absolute Error: {mae}")

    # Analyze the impact of each regressor
    regressor_coefficients = model.params['beta'][:, :len(regressor_columns)]
    regressor_impact = pd.DataFrame(data=np.abs(regressor_coefficients).mean(axis=0), columns=['Impact'], index=regressor_columns)

    # Normalize the impact to percentages
    regressor_impact['Normalized Impact (%)'] = (regressor_impact['Impact'] / regressor_impact['Impact'].sum()) * 100

    # Sort regressors by impact in descending order
    regressor_impact = regressor_impact.sort_values(by='Normalized Impact (%)', ascending=False)

    print("Regressors Impact (Normalized):")
    print(regressor_impact)

    return model, forecast, regressor_impact

# Example usage
# df = pd.read_csv('your_dataset.csv')
# model, forecast, regressor_impact = train_prophet_model(df, input_month=4, date_column='your_date_column', target_column='your_target_column')
```
